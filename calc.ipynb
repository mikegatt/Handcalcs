## Simple aluminium buckling calc

Key parameters are as below:

```python
%%render params
pi=3.14159

D = 50*mm
t = 4*mm

d_chord = 470*mm

L_cr = 6*m
E = 70*GPa
f_y = 125*N/mm**2
```

We can then calculate the properties and the buckling factors for this truss:

```python
%%render
A_chord = pi*(D/2)**2 - pi*(D/2-t)**2 # Chord Area:
I_z = 0.5 * 4 * A_chord * d_chord**2 # Second moment of area:
A = 4 * A_chord # Area:

N_cr = pi**2*E*I_z/L_cr**2 # Euler critical buckling load:
lamb_z = sqrt(A*f_y/N_cr)
alpha = 0.32
phi_z = 0.5*(1 + alpha*(lamb_z - 0) + lamb_z**2) 
chi_z = min(1/(phi_z + sqrt(phi_z**2 - lamb_z**2)),1)

N_p = (chi_z * A * f_y)/1.5 # Permissible axial buckling load
```
