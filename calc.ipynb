## Chord - 90x90x3.6

The following properties apply to this element:

```python
%%render params
b = 90*mm # Width
h = 90*mm  # Height
t = 3.6*mm  # Thickness
A = 12.3*cm**2  # Area
I_y = 152.1*cm**4  # Second moment of area
i_y = 3.5*cm  # Radius of gyration
W_ely = 33.8*cm**3  # Elastic modulus
W_ply = 39.7*cm**3  # Plastic modulus
f_y = 355*N/mm**2  # Characteristic yield stress
L = 2*m  # System Length
k = 0.9  # Effective Length factor
E = 200*GPa # Elastic modulus
gamma_M0 = 1.0 # Partial factor

```

Determine the factors for buckling and work out the design buckling load according to BS EN 1993-1-1 eq 6.47:

```python
%%render long
L_e = L * k  # Effective Length (m) = 1.8 m
N_cr = pi**2 * E * I_y / L_e**2  # Critical buckling force (kN) 
lamb_z = sqrt(A * f_y / N_cr)  # Slenderness ratio for buckling - eq 6.50
alpha_z = 0.21  # Imperfection factor - Table 6.1
phi_z = 0.5 * (1 + alpha_z * (lamb_z - 0.2) + lamb_z**2)  # Buckling reduction determination factor
chi_z = min(1 / (phi_z + sqrt(phi_z**2 - lamb_z**2)), 1)  # Buckling reduction factor - eq 6.49 
N_brd = chi_z * A * f_y # Design buckling resistance - eq 6.47 (kN) 
```

Determine bending resistance, shear resistance and tension resistance

```python
%%render long
M_cRd = W_ply * f_y / gamma_M0  # Design moment resistance (kNm) 
N_tRd = A * f_y / gamma_M0  # Design tension resistance (kN) 
V_Rd = (A / 2) * (f_y / sqrt(3)) / gamma_M0  # Design shear resistance (kN) 
```
